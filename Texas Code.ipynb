```python
import csv
with open('Texas.csv','r') as f:
    temp={}
    n=0
    data= csv.reader(f)
    for i in data:
        if len(i)>1 and -30<int(i[13].split(',')[0][:4])<50:
            if i[0] not in temp:
                n=1
                temp[i[0]]=int(i[13].split(',')[0][:4])
                
            else:
                n+=1
                temp[i[0]]=((temp[i[0]]*(n-1))+int(i[13].split(',')[0][:4]))/n

print(len(temp))
```

```python
import matplotlib.pyplot as plt 

stat = list(temp.keys())
values = list(temp.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(stat, values)
 
plt.xlabel("Stations")
plt.ylabel("Average Temperature")
plt.title("Average Temp per Stations")
plt.show()
```
